### Cleaning the stars row

The stars row contained info about the director in most of the values so to clean this field, i added director fields to the ones that didnt have and gave them values of unknown and then i split them into two rows (Director and Star) so the data can be digested properly. ### Cleaning the Votes field

To clean the votes field, i had to remove the commas and convert it to a numeric field because it was stored as a string (object) field initially.

```python
df['VOTES'] = df['VOTES'].str.replace(',', "")

df['VOTES']
```

```python
df['VOTES'] = pd.to_numeric(df['VOTES'])
```

### Filling the null values

After converting the votes field to a numeric field, i filled the null values with the mean vote number. I did same in the runtime field

```python
df['VOTES'] = df['VOTES'].fillna(df['VOTES'].mean())
df['RunTime'] = df['RunTime'].fillna(df['RunTime'].mean())

df
```

### Further cleaning

Next thing i did was to convert the votes to its appropriate data type and round the ratings and runtime field down to 2 decimal places

```python
df['VOTES'] = df['VOTES'].astype(int)
df['RATING'] = df['RATING'].round(2)
df['RunTime'] = df['RunTime'].round(2)

df
```

```python
df.info()
```

### Filling null values

I filled null values in the genre field with other and null values in the stars field with unknown

```python
df['GENRE'] = df['GENRE'].fillna('Other')
df['STARS'] = df['STARS'].fillna('Unknown')

df.info()
```

### Dropping null fields

Next thing i did was drop columns that had too many null fields

```python
df = df.dropna(thresh=20)

df.info()
```

### Cleaning the gross field

To clean the gross field i had to first remove the characters in the data and then convert it to numeric data

```python
df.loc[:, 'Gross'] = df['Gross'].str.replace('$', '')
df.loc[:, 'Gross'] = df['Gross'].str.replace('M', '')

df['Gross'].unique()
```

```python
df.loc[:, 'Gross'] = pd.to_numeric(df['Gross'])

df['Gross']
```

### Further cleaning

After converting it to numeric data, i filled the null values with the average.

```python
df.loc[:, 'Gross'] = df['Gross'].fillna(df['Gross'].mean())

df['Gross'].unique()
```

### Drop the remaining null values

```python
df = df.dropna()

df.info()
```

### Convert Start and End fields to int

```python
df['START'] = df['START'].astype(int)
df['END'] = df['END'].astype(int)

df.describe()
```

## Visualization

```python
plot_objects1 = plt.subplots(nrows=1, ncols=2, figsize=(14,5))
```

```python
# Visualizing the outliers in the RunTime field

fig, ((ax1, ax2)) = plot_objects1

sns.boxplot(df['VOTES'], ax=ax1)
ax1.set_title('Votes BoxPlot with Outliers')
sns.histplot(df['VOTES'], ax=ax2)
ax2.set_title('Votes HistPlot with Outliers')

fig
```

### Removing Outliers

IQR(Inter Quartile Range) Method

```python
q1 = df['VOTES'].quantile(0.25)
q3 = df['VOTES'].quantile(0.75)

iqr = q3 - q1
```

```python
upper_lim = q3 + (1.5 * iqr)
lower_lim = q1 - (1.5 * iqr)
```

```python
df.loc[(df['VOTES'] <= lower_lim) | (df['VOTES'] >= upper_lim)]
```

### Capping the outliers

```python
new_df = df.copy()
new_df.loc[(df['VOTES'] < lower_lim), 'VOTES'] = lower_lim
new_df.loc[(df['VOTES'] > upper_lim), 'VOTES'] = upper_lim
```

```python
plot_objects2 = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))
fig, ((ax1, ax2), (ax3, ax4)) = plot_objects2
```

```python
#Visualization grid to show the difference now that the outliers have been capped

sns.histplot(df['VOTES'], ax=ax1)
ax1.set_title('Hist Plot with outliers')
sns.histplot(new_df['VOTES'], ax=ax2)
ax2.set_title('Hist Plot without outliers')
sns.boxplot(df['VOTES'], ax=ax3)
ax3.set_title('Box Plot with outliers')
sns.boxplot(new_df['VOTES'], ax=ax4)
ax4.set_title('Box Plot without outliers')

fig
```
